```python
import glob
import json
import matplotlib.pyplot as plt
```

```python
fitness_histories = glob.glob("*/fitness_history.json")
fitness_histories
```

```python
"""Build a hashmap that points from the setting to the maximal fitness in each epoch"""
max_fitness_histories = {}
for fitness_history_file in fitness_histories:
    setting = fitness_history_file.split("26-Jan-2018_08:32:34rounds_200_num_networks_10")[1] if fitness_history_file != 'test6_dynamic_mutation_rate/fitness_history.json' else fitness_history_file
    setting = setting.split("/")[0]
    setting = setting.replace("_max_rounds_200", "")
    setting = setting.replace("_num_networks_10", "")
    with open(fitness_history_file) as f:
        content = f.read()
    fitnesses = json.loads(content)
    fitnesses = [max(x) for x in fitnesses]
    max_fitness_histories[setting] = fitnesses
```

```python
ax = plt.subplot(1,1,1)
for setting in max_fitness_histories.keys():
    ax.plot(max_fitness_histories[setting], label=setting)

ax.set_ylabel("Fitness")
ax.set_xlabel("Epoch")

# Sort the legend
handles, labels = ax.get_legend_handles_labels()
hl = sorted(zip(handles, labels), key=lambda x: max(max_fitness_histories[x[1]]), reverse=True)
handles2, labels2 = zip(*hl)
ax.legend(handles2, labels2, bbox_to_anchor=(1.1, 1.05))
plt.show()
```
